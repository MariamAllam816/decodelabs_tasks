```python
import re
import json
import random
import requests
```

```python
dataset_url = "https://raw.githubusercontent.com/jerrytigerxu/Simple-Python-Chatbot/master/intents.json"

print("Fetching dataset from GitHub...")
response = requests.get(dataset_url)

if response.status_code == 200:
    dataset = response.json()
    print("Dataset loaded successfully!\n")
else:
    print("Failed to fetch dataset. Check your internet connection.")
    exit()
```

```python
intents = {}

for intent in dataset['intents']:
    for pattern in intent['patterns']:
        clean_pattern = pattern.lower().strip()
        intents[clean_pattern] = intent['responses']
```

```python
synonyms = {
    "hey": "hello",
    "hi": "hello",
    "assist": "help"
}


fallbacks = [
    "I'm sorry, I didn't understand that. Could you rephrase?",
    "My logic engine does not recognize that command. Please type 'help'.",
    "I have no hardcoded rule for that input."
]
```

```python
chat_with_bot()
```

```python
def sanitize_input(raw_input):
    """Sanitizes text by making it lowercase and removing punctuation."""
    clean_text = raw_input.lower().strip()
    clean_text = re.sub(r'[^\w\s]', '', clean_text)
    return clean_text

def process_logic(clean_input, last_intent):
    """Atomic operation: Lookup + Context handling."""

    # 1. Synonym Normalization
    normalized_input = synonyms.get(clean_input, clean_input)

    # 2. Basic Context Tracking (Memory)
    if normalized_input == "yes" and last_intent == "options":
        return "Great! Let me know what specific system you need help with."

    # 3. Lookup & Response Generation
    matched_responses = intents.get(normalized_input)

    if matched_responses:
        return random.choice(matched_responses)
    else:
        return random.choice(fallbacks)
```

```python
def chat_with_bot():
    print("System initialized. I am your rule-based Logic Engine. Type 'exit' to end.")

    # Initialize context memory
    last_intent = None

    while True:
        user_input = input("You: ")

        # SANITIZATION
        clean_input = sanitize_input(user_input)

        # KILL COMMAND / EXIT STRATEGY
        if clean_input in ['exit', 'bye', 'quit']:
            print("Chatbot: Initiating shutdown sequence. Goodbye.")
            break

        # PROCESS & OUTPUT
        response = process_logic(clean_input, last_intent)
        print(f"Chatbot: {response}")

        # Update memory state for the next cycle
        last_intent = synonyms.get(clean_input, clean_input)
```

```python
if __name__ == "__main__":
    chat_with_bot()
```
